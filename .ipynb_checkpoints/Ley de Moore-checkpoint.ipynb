```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, Eq, dsolve
from IPython.display import display
from PIL import Image, ImageTk

t_0 = 1971

def crearEuacion(Tr_t0, tC2, TC2, tipoI, incog):
    # Definir las variables
    t, k = symbols('t k')
    T = Function('T')
    
    # Definir la ecuación diferencial
    ecuacion_diferencial = Eq(    T(t).diff(t), +k * T(t))
    pprint(ecuacion_diferencial)
    # Resolver la ecuación diferencial
    solucion1 = dsolve(ecuacion_diferencial)
    pprint(solucion1)

    # Sustituir la condición inicial en la solución específica
    solx = solve(solucion1.subs({T(t): Tr_t0, t:0}))
    pprint(solx)
    sol_cons = solx[0]
    solucion_c = solucion1.subs({'C1': sol_cons})
    print("Solucion con constante: ")
    pprint(solucion_c) # Solucion con la constante
    
    # Encontrar k en base a una condicion
    tiempo_condicion = tC2 - t_0
    condicion = solucion_c.subs({T(t):TC2 , t:tiempo_condicion})
    pprint(condicion)
    
    k_const = solve(condicion)
    soluciones_reales = [sol.evalf() for sol in k_const if sol.is_real]
    kx = soluciones_reales[0]
    kfinal = round(kx, 4)
    pprint(kfinal)
    
    solucion = solucion_c.subs({k:kfinal})
    pprint(solucion)

    # Calcular la incognita
    asintota = calcularIncognita(solucion, tipoI, incog)
    
    if tipoI == "T":
        tiempo_condicion = incog - t_0
    
    crear_grafica(solucion, tiempo_condicion, asintota)
        
def crear_grafica(solucion, rangoX, asintota_vertical):
    # Definir la variable simbólica
    t = symbols('t')

    # Definir la función T(t)
    T = Function('T')

    # Obtener la función T(t) de la solución
    T_funcion = solucion.rhs

    # Convertir la función en una función numérica
    T_numerica = lambdify(t, T_funcion, 'numpy')
    
    if asintota_vertical is not None:
        rangoX = 70
        
    # Generar valores t en el rango deseado
    valores_t = np.linspace(0, rangoX, 1000)

    # Evaluar la función numérica en los valores t
    valores_T = T_numerica(valores_t)

    # Graficar la función
    plt.plot(valores_t, valores_T, label=f'T(t) = {T_funcion}')
    # Agregar la asintota vertical si se proporciona
    if asintota_vertical is not None:
        plt.axvline(x=asintota_vertical, color='r', linestyle='--', label='t=63')
    plt.title('Gráfico de la Función T(t)')
    plt.xlabel('t')
    plt.ylabel('T(t)')
    plt.legend()
    plt.grid(True)
    
    # Guardar la gráfica como una imagen
    plt.savefig("grafica.png")

    # Cerrar la figura de Matplotlib para liberar recursos
    plt.close()

    # Mostrar la gráfica en una nueva ventana de Tkinter
    mostrar_grafica(solucion)

    #display(plt.gcf())

def calcularIncognita(solucion, tipoI, incog):
    t = symbols('t')
    T = Function('T')
    if tipoI == "T":
        tiempo_incognita = incog
        tiempo = tiempo_incognita - t_0
        solucion_incogita = solucion.subs({t:tiempo})
        pprint(solucion_incogita)
    elif tipoI == "t":
        Transistores_incognita = incog
        solucion_incogita = solucion.subs({T(t):Transistores_incognita})
        pprint(solucion_incogita)
        if Transistores_incognita == 10**12:
            print("Crear asintota")
            return 63
    return None


def mostrar_grafica(solucion):
    # Abrir la imagen con Pillow
    imagen = ImageTk.Image.open("grafica.png")

    ventana_tkinter = tk.Toplevel()
    ventana_tkinter.title("Gráfica")

    imagen_tk = ImageTk.PhotoImage(imagen)

    etiqueta_imagen = tk.Label(ventana_tkinter, image=imagen_tk)
    etiqueta_imagen.image = imagen_tk
    etiqueta_imagen.pack()
    
    ventana_tkinter.protocol("WM_DELETE_WINDOW", lambda: postVentana(ventana_tkinter, solucion))

    # Mostrar la ventana de Tkinter
    ventana_tkinter.mainloop()    

def postVentana(ventana_tkinter, solucion):
    
    
    ventana_tkinter.destroy()
```

```python
import tkinter as tk

from tkinter import *
from tkinter import messagebox

def controlarIncognita():
    return var.get() == "T"
def solucionar():
    try:
        tipo_incognita = "T" if controlarIncognita() else "t"
        T_en_t0_valor = float(texto_condicion1.get())
        T_c2_valor = float(transistores_c2.get())
        t_c2_valor = float(tiempo_c2.get())
        incognita_valor = float(texto_incognita.get()) if texto_incognita.get() else None

        crearEuacion(T_en_t0_valor, t_c2_valor, T_c2_valor, tipo_incognita, incognita_valor)

    except ValueError:
        messagebox.showerror("Error", "Ingrese solo valores numéricos.")


T_t0 = 2300
raiz = Tk()
raiz.title("Ley de Enfriamiento de Newton")

# Etiqueta de título
label_titulo = Label(raiz, text="Ley de Moore", font=("Helvetica", 16))
label_titulo.grid(row=0, column=1, pady=10)

# Explicaciones
explicaciones = Label(raiz, text="Variables: [T = Transistores, t = Tiempo]")
explicaciones.grid(row=1, column=0, padx=10, pady=10)

# Etiqueta y entrada para la condición inicial 1 (t=0)
condicion_inicial_1 = Label(raiz, text="Condición Inicial 1, T en t=0 (1971):")
condicion_inicial_1.grid(row=2, column=0, padx=10, pady=5)
texto_condicion1 = Entry(raiz)
texto_condicion1.grid(row=2, column=1, padx=10, pady=5)
texto_condicion1.insert(0, T_t0)

# Etiqueta y entradas para la condición inicial 2 (otro punto de tiempo)
condicion_inicial_2 = Label(raiz, text="Condición Inicial 2 (T y tiempo):")
condicion_inicial_2.grid(row=3, column=0, padx=10, pady=5)
transistores_c2 = Entry(raiz, width=10)
transistores_c2.grid(row=3, column=1, padx=5, pady=5)
transistores_c2.insert(0, 0)
tiempo_c2 = Entry(raiz, width=10)
tiempo_c2.grid(row=3, column=2, padx=5, pady=5)
tiempo_c2.insert(0, 0)

# Radio button para seleccionar la incógnita
var = StringVar()
var.set("T")
label_incognitas = Label(raiz, text="Seleccione la incógnita:")
label_incognitas.grid(row=4, column=0, padx=10, pady=5)
t_radio = Radiobutton(raiz, text="t  (tiempo)", variable=var, value="t", command=controlarIncognita)
t_radio.grid(row=4, column=1, pady=5)
T_radio = Radiobutton(raiz, text="T (transistores)", variable=var, value="T", command=controlarIncognita)
T_radio.grid(row=4, column=2, pady=5)

# Etiqueta y entrada para la incognita
label_incognita = Label(raiz, text="Valor de la incognita:")
label_incognita.grid(row=5, column=0, padx=10, pady=5)
texto_incognita = Entry(raiz)
texto_incognita.grid(row=5, column=1, padx=10, pady=5)
texto_incognita.insert(0, 0)

# Botón para mostrar la solucion
botonSolucion = Button(raiz, text="Mostrar Solucion", command=solucionar)
botonSolucion.grid(row=6, column=1, pady=10)

raiz.minsize(700, 300)
ancho_ventana =700
alto_ventana = 400
ancho_pantalla = raiz.winfo_screenwidth()
alto_pantalla = raiz.winfo_screenheight()
x_pos = int((ancho_pantalla / 2) - (ancho_ventana / 2))
y_pos = int((alto_pantalla / 2) - (alto_ventana / 2))
raiz.geometry(f"{ancho_ventana}x{alto_ventana}+{x_pos}+{y_pos}")

raiz.mainloop()
```
